```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data_Tanaman_Padi_Sumatera_version_1.csv.csv')
df.head()
```

```python
total_missing = df.isnull().sum().sum()
print(f"Total data missing: {total_missing}")
```

```python
df_missing = df.copy()
np.random.seed(100)
```

```python
mask_produksi = np.random.random(len(df_missing)) < 0.03
df_missing.loc[mask_produksi, 'Produksi'] = np.nan
```

```python
df_missing.info()
```

```python
mask_curah = np.random.random(len(df_missing)) < 0.04
df_missing.loc[mask_curah, 'Curah hujan'] = np.nan
```

```python
mask_kelembapan = (df_missing['Provinsi'] == 'Aceh') & (df_missing['Tahun'] <= 1998)
df_missing.loc[mask_kelembapan, 'Kelembapan'] = np.nan
```

```python
mask_suhu = (df_missing['Provinsi'] == 'Lampung') & (df_missing['Tahun'].between(2000, 2005))
df_missing.loc[mask_suhu, 'Suhu rata-rata'] = np.nan
```

```python
mask_luas = (df_missing['Provinsi'] == 'Bengkulu') & (df_missing['Tahun'] >= 2018)
df_missing.loc[mask_luas, 'Luas Panen'] = np.nan
```

```python
print("=== JUMLAH MISSING VALUE PER KOLOM ===")
print(df_missing.isnull().sum())
```

```python
total_missing = df_missing.isnull().sum().sum()
total_cells = df_missing.shape[0] * df_missing.shape[1]
print(f"\nTotal missing value: {total_missing}")
print(f"Persentase missing: {total_missing / total_cells * 100:.2f}%")
```
