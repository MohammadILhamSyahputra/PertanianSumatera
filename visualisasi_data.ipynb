```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Data_Tanaman_Padi_Sumatera_praproses.csv')

data
```

```python
plt.figure(figsize=(14, 7))
daftar_provinsi = data['Provinsi'].unique()
colors = plt.cm.tab20(range(len(daftar_provinsi)))

for i, prov in enumerate(daftar_provinsi):
    subset = data[data['Provinsi'] == prov].sort_values('Tahun')
    plt.plot(subset['Tahun'], subset['Produksi'], 
             marker='o', markersize=4, linewidth=1.5,
             label=prov, color=colors[i])

plt.xlabel('Tahun', fontsize=12)
plt.ylabel('Produksi (juta ton)', fontsize=12)
plt.title('Tren Produksi Padi Seluruh Provinsi di Sumatera (1993-2020)', fontsize=14)
plt.legend(title="Provinsi", bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=9)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
rata_produksi = data.groupby('Provinsi')['Produksi'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
warna = plt.cm.Greens(np.linspace(0.3, 0.9, len(rata_produksi)))
plt.bar(rata_produksi.index, rata_produksi.values, color=warna)
plt.xlabel('Provinsi')
plt.ylabel('Rata-rata Produksi (juta ton)')
plt.title('Rata-rata Produksi Padi per Provinsi')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
agregat = data.groupby('Provinsi').agg({
    'Produksi': 'mean',
    'Luas Panen': 'mean',
    'Curah hujan': 'mean'
}).sort_values('Produksi', ascending=False)
fig, ax = plt.subplots(figsize=(12, 6))
(agregat['Produksi'] / 1000000).plot(
    kind='bar', 
    ax=ax, 
    color='blue', 
    alpha=0.7, 
    label='Rata-rata Produksi (Juta Ton)'
)
ax.set_ylabel('Rata-rata Produksi (Juta Ton)', fontsize=12)
ax.set_xlabel('Provinsi', fontsize=12)
ax2 = ax.twinx()
agregat['Curah hujan'].plot(
    kind='line', 
    ax=ax2, 
    color='red', 
    marker='o', 
    linewidth=2, 
    label='Rata-rata Curah Hujan (mm)'
)
ax2.set_ylabel('Rata-rata Curah Hujan ($mm$)', fontsize=12)
plt.title('Perbandingan Rata-rata Produksi dan Rata-rata Curah Hujan per Provinsi', fontsize=14)
ax.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.3) # Tambahan grid halus agar lebih rapi
plt.tight_layout()
plt.show()
```

```python
def plot_pie_tahun(tahun):
    data_tahun = data[data['Tahun'] == tahun].copy()
    if data_tahun.empty:
        print(f"Tidak ada data untuk tahun {tahun}")
        return
    idx_max = data_tahun['Produksi'].idxmax()
    explode = [0.1 if i == idx_max else 0 for i in data_tahun.index]
    
    plt.figure(figsize=(8, 8))
    plt.pie(data_tahun['Produksi'], labels=data_tahun['Provinsi'], autopct='%1.1f%%',
            startangle=140, explode=explode,
            colors=plt.cm.Paired(np.linspace(0, 1, len(data_tahun))))
    plt.title(f'Proporsi Produksi Padi Sumatera ({tahun})', pad=20, fontsize=14)
    plt.tight_layout()
    plt.show()

plot_pie_tahun(1993)
plot_pie_tahun(2020)
```

```python
data['Produktivitas'] = data['Produksi'] / data['Luas Panen']
rata_prod = data.groupby('Provinsi')['Produktivitas'].mean().sort_values()
plt.figure(figsize=(12, 7))
ax = rata_prod.plot(kind='barh', color="#ff9b04", alpha=0.8)

plt.xlabel('Produktivitas (Ton per Hektar)', fontsize=12)
plt.ylabel('Provinsi', fontsize=12)
plt.title('Rata-rata Produktivitas Padi per Hektar per Provinsi (1993-2020)', fontsize=14, pad=15)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.xlim(0, rata_prod.max() + 1)
plt.tight_layout()
plt.show()
```
