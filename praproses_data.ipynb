```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
for prov in data['Provinsi'].unique():
    median_prov = data[data['Provinsi'] == prov]['Produksi'].median()
    data.loc[(data['Provinsi'] == prov) & (data['Produksi'].isnull()), 'Produksi'] = median_prov
```

```python
Q1 = data['Produksi'].quantile(0.25)
Q3 = data['Produksi'].quantile(0.75)
IQR = Q3 - Q1
upper = Q3 + 1.5 * IQR
data['Produksi'] = np.where(data['Produksi'] > upper, upper, data['Produksi'])
data['Produksi']
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data['Produksi'])
plt.title('Boxplot Produksi Padi (Setelah Penanganan Outlier)')
plt.ylabel('Produksi (ton)')
plt.show()
```

```python
# Cek apakah ada luas panen di bawah 100.000 hektar
data_dibawah_100rb = len(data[data['Luas Panen'] < 100000])
print(f"Data di bawah 100.000 hektar: {data_dibawah_100rb}")
```

```python
# Menyimpan hasil filter ke variabel baru
data_kecil = data[data['Luas Panen'] < 100000]

# Menampilkan datanya
data_kecil
```

```python
# Cek apakah ada curah hujan di atas 6.000 mm
data_diatas_6rb = len(data[data['Curah hujan'] > 6000])
print(f"Data di atas 6.000 mm: {data_diatas_6rb}")
```

```python
# Menyimpan hasil filter ke variabel baru
data_kecil = data[data['Curah hujan'] > 6000]

# Menampilkan datanya
data_kecil
```

```python
data['Curah hujan'] = data['Curah hujan'].replace(51110.0, 5111.0)
data['Curah hujan'] = data['Curah hujan'].replace(44869.5, 4486.95)
data['Curah hujan'] = data['Curah hujan'].replace(22211.3, 2221.13)
data['Curah hujan']
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data['Curah hujan'])
plt.title('Boxplot Curah Hujan (Setelah Penanganan Outlier)')
plt.ylabel('Curah Hujan (mm)')
plt.show()
```

```python
# 4. KELEMBAPAN - Ganti nilai di bawah batas bawah dengan batas bawah
Q1 = data['Kelembapan'].quantile(0.25)
Q3 = data['Kelembapan'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
data['Kelembapan'] = np.where(data['Kelembapan'] < lower, lower, data['Kelembapan'])
data['Kelembapan']
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data['Kelembapan'])
plt.title('Boxplot Kelembapan (Setelah Penanganan Outlier)')
plt.ylabel('Kelembapan (%)')
plt.show()
```

```python
# 5. SUHU RATA-RATA - Ganti nilai di luar batas (bawah & atas)
Q1 = data['Suhu rata-rata'].quantile(0.25)
Q3 = data['Suhu rata-rata'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
data['Suhu rata-rata'] = np.where(data['Suhu rata-rata'] < lower, lower, data['Suhu rata-rata'])
data['Suhu rata-rata'] = np.where(data['Suhu rata-rata'] > upper, upper, data['Suhu rata-rata'])
data['Suhu rata-rata']
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(data['Suhu rata-rata'])
plt.title('Boxplot Suhu Rata-Rata (Setelah Penanganan Outlier)')
plt.ylabel('Suhu Rata-Rata (°C)')
plt.show()
```
