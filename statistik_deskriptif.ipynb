```python
print("\n=== JUMLAH DATA PER PROVINSI ===")
jumlah_per_provinsi = data.groupby('Provinsi')['Tahun'].count()
print(jumlah_per_provinsi)
```

```python
print("\n=== RATA-RATA PRODUKSI PER PROVINSI ===")
rata_produksi_per_provinsi = data.groupby('Provinsi')['Produksi'].mean().sort_values(ascending=False)
print(rata_produksi_per_provinsi)
```

```python
print("\n=== STATISTIK PRODUKSI PER PROVINSI (describe groupby) ===")
print(data.groupby('Provinsi')['Produksi'].describe())
```
